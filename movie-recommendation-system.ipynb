# *[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'we have to convert list of dict in the single list of names like['act','adv',afan','scific']*

```python
import ast # this convert list string of list into a list
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L   
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert2(obj):
    L=[]
    a=0
    for i in ast.literal_eval(obj):
        if a !=3:
           L.append(i['name'])
           a+=1
        else:
           break
    return L   
# here we want only first three actor so we take a==0 to a!=3
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
import ast

def fetch_director(obj):
    L = []
    try:
        # If obj is a string, convert it
        crew_list = ast.literal_eval(obj) if isinstance(obj, str) else obj
        
        for i in crew_list:
            if i.get('job') == 'Director':
                L.append(i.get('name'))
    except (ValueError, SyntaxError, TypeError):
        pass
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

# *Steammers*

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stem (text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  # convert list into string
    
```

```python
stem('loved loving dancing dance') # this is the example of above function
```

```python
new_movies['tag'].apply(stem)
```

# *Now we can calculate the cosine similarity between two vectors for distance*

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]#distance of movie 1 with each movie
```

# *form main function*

```python
# we try when movie given to us we calculate the top 5 maximum distnce for that movie from other movies 
# we can give that 5 maximum distance(cosine similarity should be max) movies as recommend movie for that given movie
```

```python
# if we sort the distances than index should be chnage so we use enumerate fumction here
```

```python
sorted(list(enumerate(similarity[0])),reverse=True) # here we can see each distance assign a index
# but the problem is it sort according to index
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] 
# to resolve the above issue we write code lambda and select first five distances
```

```python
def recommend (movie):
    movie_index=new_movies[new_movies['title']==movie].index[0] # here we can calculate the index of given movie
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] 

    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
       
```

```python
recommend('Avatar') # here we want name of movies
```
